There are so much unnecessary columns in our dataframe. So we can create new dataframe.

```python
gof_df = df.loc[1: ,["name", "year", "attacker_king","attacker_size", "defender_king","defender_size",
                     "attacker_commander", "defender_commander", "battle_type","location","region"]]
```

```python
gof_df
```

```python
war_year = gof_df["year"].value_counts()
```

```python
# Number of battles by years
war_year
```

```python
plt.style.use("seaborn")

plt.bar(gof_df["year"],gof_df["name"])

plt.xlabel("years")
plt.ylabel("name of battle")
plt.title("battles by years")

plt.show()


# apparently, battles have increased constantly 
```

We want to see total number of soldier.

```python
gof_df["total_soldier"] = gof_df["attacker_size"] + gof_df["defender_size"]
```

```python
gof_df.drop(columns = ["attacker_size", "defender_size"], inplace = True)
```

```python
gof_df
```

```python
plt.style.use("seaborn-pastel")

plt.barh(gof_df["name"],gof_df["total_soldier"])

plt.xlabel("number of soldier")
plt.ylabel("name of battles")
plt.title("number of soldier by battles")

plt.show()

#battle of castle black baya insan savaÅŸmÄ±ÅŸ :D
```

```python
# We can see the battle that the fewest people

gof_df.loc[ : , ["name", "total_soldier"]].apply(lambda x: x.min())
```

We want to see the battles Robb Stark had

```python
filt = (gof_df["attacker_king"] == "Robb Stark") | (gof_df["defender_king"] == "Robb Stark") 
```

```python
filt
```

```python
robb_df = gof_df.loc[filt,["name","year", "defender_king","attacker_king"]]
```

```python
# This dataframe show us the battles Robb Stark had

robb_df
```
